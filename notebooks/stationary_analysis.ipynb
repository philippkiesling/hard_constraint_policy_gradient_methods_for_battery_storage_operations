```python
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

test_results = adfuller(df["intraday_15min"][-500])
test_results
```

```python
print(f"p-value: {test_results[1]}")
```

p-value is well bellow 0.05, so we can reject the null hypothesis and say that the data is stationary.
